```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = statdf[['max_LC', 'skew', 'CV_LC' , 'mad', "aslope", "bslope", "median", "time"]]
y = np.ravel(statdf[['startype']])

X_trans = scaler.fit_transform(X)
#y.reshape(y[0].shape)

X_train, X_test, y_train, y_test = train_test_split(X_trans,y,test_size = 0.2, random_state=201)

print(np.max(X_trans)) 
```

```python
print(len(statdf['startype']))

statdf0 = statdf[statdf["startype"]== 0]
statdf1 = statdf[statdf["startype"]== 1]
statdf2 = statdf[statdf["startype"]== 2]
print(statdf[statdf['startype'] ==2])

#print(len(statdf0), len(statdf1), len(statdf2))

min_size = min(len(statdf0), len(statdf1), len(statdf2)) 

statdf0 = statdf0.sample(n = min_size)
statdf2 = statdf2.sample(n = min_size)
statdf1 = statdf1.sample(n = min_size)
statdf = pd.concat([statdf0, statdf1, statdf2])

stat_train, stat_test, ystat_train, ystat_test = train_test_split(statdf
    [["max_LC","skew","CV_LC", "mad", "aslope", "bslope", "median", "time"]],
    statdf["startype"],test_size = 0.2, random_state=5)

print(len(statdf1))
```

```python
from sklearn import tree
#clf = MLPClassifier(solver = 'lbfgs', alpha = 1e-5,
#                     hidden_layer_sizes = (5), 
#                     random_state = 1, 
#                     max_iter = 2000)

# lbfgs default
#clf = RandomForestClassifier(n_estimators = 100)
clf = DecisionTreeClassifier()

#clf = SVC()
#clf = KNeighborsClassifier(n_neighbors = 3)
#clf = KMeans(n_clusters = 3, random_state = 0)
clf.fit(stat_train, ystat_train)

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(16,6))
tree.plot_tree(clf, max_depth=3, ax=ax, fontsize=8,feature_names=
               ["max_LC", "skew", "CV_LC", "mad", "aslope", "bslope", "median", "time"], class_names=
               ["Arnett", "Magnetar", "CSM"])

plt.show()
print(clf.tree_.max_depth)
# plt.savefig("tree_example.jpeg")
#plt.show()
#print(X_test)
y_pred = clf.predict(stat_test)
#print(y_pred)

print('Accuracy:', metrics.accuracy_score(y_pred,ystat_test))
```
